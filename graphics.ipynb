```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bipedal Walker

## Reward
df_bipedal_baseline_reward = pd.read_csv('./data/results_ppo-baseline-bipedalwalker-hardcore-long.csv')
df_bipedal_curriculum_reward = pd.read_csv('./data/results_ppo-curriculum-bipedalwalker-long.csv')

## Entropy
df_bipedal_baseline_entropy = pd.read_csv('./data/results_ppo-baseline-bipedalwalker-hardcore-long-entropy.csv')
df_bipedal_curriculum_entropy = pd.read_csv('./data/results_ppo-curriculum-bipedalwalker-long-entropy.csv')

# Lunar lander

## Reward
df_lunar_baseline_reward = pd.read_csv('./data/results_ppo-baseline-lunarlander-long.csv')
df_lunar_curriculum_reward = pd.read_csv('./data/results_ppo-curriculum-lunarlander-long.csv')

## Entropy
df_lunar_baseline_entropy = pd.read_csv('./data/results_ppo-baseline-lunarlander-long-entropy.csv')
df_lunar_curriculum_entropy = pd.read_csv('./data/results_ppo-curriculum-lunarlander-long-entropy.csv')
```

```python
# Plotting

# plt.style.use('dark_background')
# plt.rcParams['axes.facecolor'] = '#26262e' 

def plot_reward(df_baseline, df_curriculum, title):
    plt.figure(figsize=(10, 5))
    plt.plot(df_baseline['Step'], df_baseline['Value'], label='Baseline')#, color='blue')
    plt.plot(df_curriculum['Step'], df_curriculum['Value'], label='Curriculum')#, color='orange')
    # Draw a vertical line at x=1_000_000
    plt.axvline(x=1_000_000, color='red', linestyle='--', label='Task 1')
    # Draw a vertical line at x=2_000_000
    plt.axvline(x=2_000_000, color='green', linestyle='--', label='Task 2')
    plt.title(title)
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.legend()
    plt.grid()
    plt.show()
    
def plot_entropy(df_baseline, df_curriculum, title):
    plt.figure(figsize=(10, 5))
    plt.plot(df_baseline['Step'], df_baseline['Value'], label='Baseline')#, color='blue')
    plt.plot(df_curriculum['Step'], df_curriculum['Value'], label='Curriculum')#, color='orange')
    # Draw a vertical line at x=1_000_000
    plt.axvline(x=1_000_000, color='red', linestyle='--', label='Task 1')
    # Draw a vertical line at x=2_000_000
    plt.axvline(x=2_000_000, color='green', linestyle='--', label='Task 2')
    plt.title(title)
    plt.xlabel('Episode')
    plt.ylabel('Entropy')
    plt.legend()
    plt.grid()
    plt.show()

# Bipedal Walker
plot_reward(df_bipedal_baseline_reward, df_bipedal_curriculum_reward, 'Bipedal Walker - Reward')
plot_entropy(df_bipedal_baseline_entropy, df_bipedal_curriculum_entropy, 'Bipedal Walker - Entropy')
```
